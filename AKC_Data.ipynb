# Standard Life Expectancy 

```python
dogs['Life Expectancy'][2]
```

```python
def lifespan(row, min):
    ''' 
        Lifespans are formated 'x-y years', 'x-years' or ''
        Return the min x if min is true
        Return the max y if min is false
        Return X if formatted 'x-years'
        return 'Unknown' if ''
    '''
    i = re.findall(r'\b\d+\b', row)
    if len(i) == 0:
        return 'Unknown'
    if len(i) == 1:
        return i[0]
    if min:
        return i[0]
    else:
        return  i[1]


dogs['Min Life Expectancy'] = dogs['Life Expectancy'].apply(lifespan, min = True)
dogs['Max Life Expectancy'] = dogs['Life Expectancy'].apply(lifespan, min = False)
```

# Getting Dog Rankings by Year


```python
df_2023 = pd.read_csv('Data/Most_Popular/top_200_2023.csv')
df_2022 = pd.read_csv('Data/Most_Popular/top_200_2022.csv')
df_2021 = pd.read_csv('Data/Most_Popular/top_200_2021.csv')
df_2020 = pd.read_csv('Data/Most_Popular/top_200_2020.csv')
df_2019 = pd.read_csv('Data/Most_Popular/top_200_2019.csv')
```

```python
rankings = [df_2023, df_2022, df_2021, df_2020, df_2019]
```

```python
unique_breeds = set()

def standardize_breed_name(breed):
    #check if it is a string just in case
    if isinstance(breed, str):
        # remove plural "s" at the end for breeds like "Pointers", "Retrievers", "Spaniels", etc.
        breed = re.sub(r'\s?(s)$', '', breed)  
        breed = re.sub(r's\s$', '', breed) #really enforce no s
        
        # handle the case where a breed has the format "Dog (Specific Type)"
        match = re.match(r'([a-zA-Z\s]+)\s?\(([^)]+)\)', breed)
        if match:
            # reformat to "Specific Type Dog"
            return f"{match.group(2)} {match.group(1)}"
        
        # replace non-breaking space with regular space
        breed = breed.replace('\xa0', ' ')
        
        #aggressively getting rid of the s does hurt the malligator, fix her
        if breed == 'Belgian Malinoi':
            breed = 'Belgian Malinois'
    
    return breed

# Loop over each DataFrame in the rankings
for year in rankings:  
    # Apply the breed standardization to the 'Breed' column
    year['Breed'] = year['Breed'].apply(standardize_breed_name)

# Update the unique_breeds set with unique breed names from each DataFrame
for year in rankings:
    unique_breeds.update(year['Breed'].unique())

# Convert the set to a list
unique_breeds = list(unique_breeds)

# Optional: Print out the first few standardized breeds to verify
print(unique_breeds[:10])
```

```python
unique_breeds
```

```python
final_list = []
dog_yearly_ranking =  pd.DataFrame()

for breed in unique_breeds:
    breed_rank = [breed]  # Start the list with the breed name
        
    # Loop through the dataframes and get the rank for each breed
    for year in rankings:  # Loop from 2023 to 2019
        
        # Get the rank for the breed in the current DataFrame
        rank = year[year['Breed'] == breed]['Rank'].values
        if len(rank) > 0:
            breed_rank.append(rank[0])
        else:
            breed_rank.append("NA")  # In case the breed is not found in the year
        
    final_list.append(breed_rank)

# Final result
print(final_list)
```

## Now that we have a list, make a dataframe! 

```python
breeds = [row[0] for row in final_list]  # Extract breeds
year_data = [row[1:] for row in final_list] # Extract year data

df = pd.DataFrame(year_data, index=breeds, columns=['2023', '2022', '2021', '2020', '2019'])

df_long = df.stack().reset_index()
df_long.columns = ['Breed', 'Year', 'Rank']

print(df_long)
```
